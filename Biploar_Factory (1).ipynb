```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import time
```

# opening head page

```python
url = "https://www.dailymail.co.uk"

r = requests.get(url)
print(r.status_code)
coverpage = r.content

# Soup creation
#soup1 = BeautifulSoup(coverpage3, 'html5lib')
#cover = soup1.find_all('a', class_='home')

```

# All link of nav bar

```python
soup1 = BeautifulSoup(coverpage, 'html5lib')
cover = soup1.find_all(class_='link-gr6ox linkro-ccox')
```

```python
count=0
ch=[]
all_link=[]
for item in cover:
  if count<13:
    title = item.find('a').attrs['href']
    url="https://www.dailymail.co.uk" + title
    print(url)
    ch.append(url)
    count+=1
```

```python
listtitles=[]
new=[]
listrela=[]
c_count=[]


```

# Fetching all links on each page of website

```python
base="https://www.dailymail.co.uk"
for i in ch:

  r1 = requests.get(i)
  print(r1.status_code)
  coverpage1 = r1.content

  # Soup creation
  soup1 = BeautifulSoup(coverpage1, 'html5lib')
  for ultag in soup1.find_all('ul', {'class': 'float-l'}):
  # News identification
    for litag in ultag.find_all('li'):
      lin=litag.a['href']
      if 'http' in lin:
        continue
      else:
        lin=base+lin
        r2 = requests.get(lin)
        #print(r4.status_code)
        coverpage2 = r2.content

        # Soup creation
        soup1 = BeautifulSoup(coverpage2, 'html5lib')

        # News identification
        coverpage_news2 = soup1.find_all('h2', class_='linkro-darkred')
        len(coverpage_news2)
        coverpage_comments = soup1.find_all('span', class_='readerCommentNo bold')
        len(coverpage_comments)
        for i in range(len(coverpage_news2)):

          title = coverpage_news2[i].find('a').attrs['href']
          listtitles.append(title)
        for i in range(len(coverpage_news2)):

          head = coverpage_news2[i].find('a').get_text()
          new.append(head)
        rela=soup1.find_all(class_='readerCommentNo bold')
        for rl in rela:
          entry=rl.attrs['rel']
          listrela.append(entry)
        
        posts=soup1.find_all(class_ = "readerCommentNo bold")
        for post in posts:
          txt=post.text
          c_count.append(txt)
```

```python
list_id=[]
for i in listtitles:
  num=re.sub('\D', '',i)
  n1=num[:7]
  list_id.append(n1)
list_id
```

```python
for i in range(len(c_count)):
  if 'k' in c_count[i]:
    c_count[i]= c_count[i].replace('k','')
    c_count[i]=float(c_count[i])
    c_count[i]=c_count[i]*1000
    c_count[i]=int(c_count[i])
  else:

    c_count[i]=int(c_count[i])
```

```python
u_id=[]
headline=[]
count=[]
for i in list_id:
    if(i in listrela):
      ind=list_id.index(i)
      in1=listrela.index(i)
      u_id.append(i)
      headline.append(new[ind])
      count.append(c_count[in1])
```

```python
len(u_id),len(headline),len(count)
```

Percentile to divide output in 4 sections

```python
num1=int(np.percentile(count,25))
num2=int(np.percentile(count,50))
num3=int(np.percentile(count,75))
num1,num2,num3
```

```python
h=[]
#total=0
for i in count:
  if i> num3:
    h.append(3)
  elif i< num3 and i > num2:
    h.append(2)
  elif i<num2 and i>num1:
    h.append(1)
  else:
    h.append(0)
#total
```

# Dataset

# Splitting of Data

```python
X=df_show_info['comments']
y=df_show_info['viral']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lrc = LogisticRegression()
lrc.fit(X_train.values.reshape(-1,1), y_train)
```

```python
lrc_pred = lrc.predict(X_test.values.reshape(-1,1))
```

# Accuracy_Score

```python
print(accuracy_score(y_test,lrc_pred))
```
